<font color='grey'>Understanding the Data</font>

This dataset contains **2500 rows** of **hospitalization data** that an **insurance company** is analysing, where the **Insurance charges** are given against the following **attributes of the insured:** Gender, Age (Year_Birth), Marital Status, BMI, Region, etc.
<br><br>

**->Insured identifier**:
- `Insured ID` - Insurance contractor identifier.


**->Sociodemographic data**:
- `Year_Birth` - Insurance contractor year of birth.
- `Gender` - Insurance contractor gender: "female" or "male".
- `Region` - The beneficiary's residential area in the US: "northeast" or "southeast"or "southwest" or "northwest".
- `Marital Status` - Insurance contractor marital status.
- `Smoker` - "Smoker" or "Non - smoker".
- `Income` - Insurance contractor income.
- `BMI` - Body mass index, providing an understanding of body, weights that are relatively high or low relative to height, objective index of body weight (kg / m2).
- `BSA` - Body surface area.


**->Hospitalization data**:
- `Dt_Hospitalization` - Insurance contractor hospitalization date.
- `Insured_Satisfaction` - Insured satisfaction regarding insurance assistance/services covered during hospitalization.
- `Expenses` - Individual medical costs billed by health insurance.
- Expenses in percentage (%) by categories: `Treatment` / `Medication` / `Medical_Assistance` / `Exams` / `Ambulance_Transport`.
- Expenses coverage percentage = (`Insurance_Coverage` / `Patient_Coverage`).


**->Firmographic Data**
- `Health_Plan_Option` - Insurance health plan chosen by the insurance contractor.
- `Upgrade_Health_Plan` - If insurance contractor updated the health plan or not.
<br><br>

**Step 11 -**  Import seaborn and define as alias the name `sns`.

**Step 12 -** Using the seaborn method `boxplot()`, create a bivariate plot by defining the parameters:
- `x = 'Smoker'`

```python
sns.boxplot(x='Smoker', y='Expenses', data=data, hue='Smoker', palette=["yellowgreen", "lightgrey"], legend=False)
```

**Step 13 -**  Using the seaborn method `countplot()`, create a new plot by defining the parameters:

- `hue = 'Gender'`

```python
sns.countplot(y = 'Region', hue = 'Gender', data = data, palette='dark:yellowgreen')
```

**Step 14 -**  Validate if the dependent variable `Upgrade_Health_Plan` is balanced by using the method `sns.countplot`.

```python
sns.countplot(x='Upgrade_Health_Plan', data=data, hue='Upgrade_Health_Plan', palette=["yellowgreen","lightgrey"], legend=False)
```

**Step 15 -**  We can change several aesthetic aspects in out plot by using `matplotlib.pyplot` functions.
See more in <a href='https://matplotlib.org/stable/api/pyplot_summary.html'> here </a>. <br>
In the following cell we change the limits of the axis, the label in ticks in axis, the title, the labels of the axis and the legend, but it is possible to change much more...

**Step 15.1 -**  Using the seaborn method `scatterplot()`, **create a new plot** by defining the parameters:
- `x = 'Income'`
- `y = 'Expenses'`
- `hue = 'Insured_Satisfaction'`
- `data = 'data'`

**Step 15.2 -**  Define the **limits of the y axis** by calling the matplotlib.pyplot method `ylim()` and define the parameters:
- `bottom = -500`
- `top = None`


**Step 15.3 -** Define the **limits of the x** axis by calling the matplotlib.pyplot method `xlim()` and define the parameters:
- `left = 0`
- `right = None`


**Step 15.4 -**  Define the **ticks in x axis** using `matplotlib.pyplot.xticks(start, stop, step)`. <br>
Note: In this case we are going to call a function from the library numpy named arange, that returns evenly spaced values within a given interval. So we need to import also the numpy library (as np) to the notebook.


**Step 15.5 -**  Define the **title** using `matplotlib.pyplot.title()` and the following parameters:
- `label = 'Income vs Expenses vs Satisfaction'`
- `fontsize= 16`
- `color = 'yellowgreen'`


**Step 15.6 -**  Define the **position of the legend**using `matplotlib.pyplot.legend()` and the following parameter:
- `loc = lower left`


**Step 15.7 -**  Define the **label for x axis** using `matplotlib.pyplot.xlabel()` and the following parameter:
- `xlabel = 'Income'`


**Step 15.8 -**  Define the **label for y axis** using `matplotlib.pyplot.ylabel()` and the following parameter:
- `ylabel = 'Expenses'`



```python
# Step 15.1
sns.scatterplot(x = 'Income', y = 'Expenses', hue = 'Insured_Satisfaction', data = data, palette = ['darkred','red', 'orange', 'yellow', 'yellowgreen', 'greenyellow'])

# Step 15.2
plt.ylim(bottom = -500, top = None)

# Step 15.3
plt.xlim(left = 0, right = None)

# Step 15.4
plt.xticks(np.arange(30000,140000,30000))

# Step 15.5
plt.title(label = 'Income vs Expenses vs Satisfaction', fontsize= 16, color = 'yellowgreen')

# Step 15.6
plt.legend(loc = 'lower left')

# Step 15.7
plt.xlabel(xlabel = 'Income')

# Step 15.8
plt.ylabel(ylabel = 'Expenses')
```

**Step 16 -** If we want to **visualize pairwise relationships**, the `pairplot` method is indicated for that purpose. This will give us an **histogram** for each of the variables and a **scatterplot** for each pair of variables.

**Step 16.1 -**  Create a new dataset names `data_num` using the `copy()` pandas method with the variables 'Expenses', 'Medication', 'Treatment', 'Insurance_Coverage'.

```python
data_num = data[['Expenses', 'Medication', 'Treatment', 'Insurance_Coverage']].copy()
```

**Step 16.2 -**  Using the seaborn method `pairplot()`, create a new plot by defining the parameters:
- `data = 'data_num'`

```python
sns.pairplot(data = data_num, plot_kws = {'color':'yellowgreen'}, diag_kws = {'color':'darkgreen'})
```
